```python
report = open('../summary/summary.txt', 'r')
content = report.read()
report.close()
```

```python
lines = content.split("\n\n")[:-1]
map = {}
key = 1
for line in lines:
    parts = line.split()
    map[key] = []
    for i in range(len(parts)):
        if parts[i] == "threads:":
            t = float(parts[i + 1])
            map[key].append(t)
    key += 1
```

```python
import matplotlib.pyplot as plt
import sys

data = map.get(1)
num_threads = [1, 2, 4, 8, 16, 32, 64, 128]

min_time = sys.float_info.max
thread = 0
for i in range(len(data)):
    if data[i] < min_time:
        min_time = data[i]
        thread = num_threads[i]

print("The number of threads with the lowest runtime is " + str(min_time) + " is: " + str(thread))

x_values = range(1, len(data) + 1)

plt.plot(num_threads, data, marker='o')
plt.xlabel('Number of Threads')
plt.ylabel('Average Elapsed Time (miliseconds)')
plt.title('Number of Threads and its average runtime')
plt.grid(True)
plt.xscale('log', base = 2)
plt.xticks(num_threads, labels = num_threads)

plt.show()
```
